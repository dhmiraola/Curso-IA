```python
import pandas as pd

data = pd.read_csv('housing.csv')

data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data['ocean_proximity'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data['ocean_proximity'] = encoder.fit_transform(data['ocean_proximity'])
data.head()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Separar los datos
X = data.drop('median_house_value', axis=1)
y = data['median_house_value']

# Dividir los datos
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)

# Normalizar
scaler = StandardScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)

# Crear el modelo de Red Neuronal
model = Sequential([
    Input(shape=(X_train_normalized.shape[1],)),
    Dense(128, activation='relu'),
    Dropout(0.2),  # previene sobreajuste
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='linear')  # 1 neurona cuando es salida binaria o cuando es regresion
])

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='mean_squared_error',
    metrics=['mean_squared_error']
)

# Entrenamiento
model.fit(X_train_normalized, y_train, epochs=100, batch_size=16, verbose=1)


```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Predicciones del modelo
y_pred = model.predict(X_test_normalized).flatten()

# Cálculo de métricas
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# Mostrar resultados
print("📊 Métricas del modelo de regresión:")
print(f"MAE  (Error absoluto medio):     {mae:.2f}")
print(f"RMSE (Raíz del error cuadrático): {rmse:.2f}")
print(f"R²   (Coeficiente de determinación): {r2:.3f}")
```
